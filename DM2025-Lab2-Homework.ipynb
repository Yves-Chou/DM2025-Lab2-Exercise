#### 1.3.2 Text Cleaning

```python
!pip install emoji
```

```python
import re
import emoji

# Clean Emoji
def normalize_emoji(text: str) -> str:
  text = emoji.demojize(text)
  text = re.sub(r":([a-z0-9_]+):", r" EMO_\1 ", text)
  return text

# Clean Text
def clean_text(text: str) -> str:
  # Emoji
  text = normalize_emoji(text)

  # Lower
  text = text.lower()

  # HTML entity
  text = text.replace("&amp;", " and ")

  # Remove urls
  text = re.sub(r"http\S+|www\.\S+", " ", text)

  # Remove @mentions
  text = re.sub(r"@\w+", " ", text)

  # Keep hashtag word, drop '#'
  text = re.sub(r"#(\w+)", r"\1", text)

  # Normalize placeholders like [NAME], [RELIGION]
  text = re.sub(r"\[(name|religion)\]", r" \1 ", text, flags=re.IGNORECASE)

  # remove extra whitespace
  text = re.sub(r"\s+", " ", text).strip()

  return text
```

```python
for df_ in (train_df, test_df):
    df_["clean_text"] = df_["text"].apply(clean_text)
```

#### 1.3.3 Tokenization

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")


# Stopword
EN_STOPWORDS = set(stopwords.words("english"))

# Negative Words
NEGATIVE_WORDS = {
    "no", "nor", "not",
    "don't", "didn't", "doesn't", "can't", "couldn't",
    "won't", "wouldn't", "shouldn't",
    "isn't", "wasn't", "aren't", "weren't",
    "ain't", "n't"
}
EN_STOPWORDS = EN_STOPWORDS - NEGATIVE_WORDS

def tokenize_from_clean(text: str):
  tokens = word_tokenize(text)

  cleaned_tokens = [
      t for t in tokens
      if t.lower() not in EN_STOPWORDS and len(t) > 1
  ]
  return cleaned_tokens
```

```python
for df_ in (train_df, test_df):
    df_["clean_text"] = df_["text"].apply(clean_text)
    df_["tokens"] = df_["clean_text"].apply(tokenize_from_clean)
```

## 2. Feature Engineering Steps: BERT

```python
!pip install -q transformers accelerate
```

```python
import torch
from transformers import AutoTokenizer, AutoModel
import numpy as np

# BERT base
MODEL_NAME = "bert-base-uncased"

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
bert_model = AutoModel.from_pretrained(MODEL_NAME)
bert_model.to(device)
bert_model.eval()

def encode_with_bert(texts, batch_size: int = 32, max_length: int = 128):
  all_embeddings = []

  for start in range(0, len(texts), batch_size):
    batch_texts = texts[start:start + batch_size]

    enc = tokenizer(
        batch_texts,
        padding=True,
        truncation=True,
        max_length=max_length,
        return_tensors="pt"
    )

    input_ids = enc["input_ids"].to(device)
    attention_mask = enc["attention_mask"].to(device)

    with torch.no_grad():
      outputs = bert_model(
          input_ids=input_ids,
          attention_mask=attention_mask
      )
      last_hidden_state = outputs.last_hidden_state  # [batch, seq_len, 768]

    # Mean pooling with attention mask ===
    mask = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
    sum_embeddings = (last_hidden_state * mask).sum(dim=1)
    sum_mask = mask.sum(dim=1).clamp(min=1e-9)
    mean_embeddings = sum_embeddings / sum_mask  # [batch, 768]

    all_embeddings.append(mean_embeddings.cpu())

  all_embeddings = torch.cat(all_embeddings, dim=0)
  return all_embeddings.numpy()

# Text
train_texts = train_df["clean_text"].tolist()
test_texts  = test_df["clean_text"].tolist()

print("Encoding train with BERT...")
X_train = encode_with_bert(train_texts, batch_size=32, max_length=128)

print("Encoding test with BERT...")
X_test = encode_with_bert(test_texts, batch_size=32, max_length=128)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Logistic Regression
Cs = [0.5, 1.0, 2.0, 4.0]
best_lr = None
best_f1 = -1
best_C = None

for C in Cs:
  lr_tmp = LogisticRegression(
      C=C,
      solver="lbfgs",
      multi_class="auto",
      max_iter=1000,
      class_weight=None,
      n_jobs=-1,
  )
  lr_tmp.fit(X_tr_bal, y_tr_bal)
  y_val_pred = lr_tmp.predict(X_val)
  f1 = f1_score(y_val, y_val_pred, average="macro")
  print(f"LR_bal (C={C}) Val Macro-F1: {f1:.4f}")

  if f1 > best_f1:
    best_f1 = f1
    best_lr = lr_tmp
    best_C = C

print("Best C:", best_C, "Best Val Macro-F1:", best_f1)
```

```python
# Fit
lr_final = LogisticRegression(
    C=best_C,
    solver="lbfgs",
    multi_class="auto",
    max_iter=1000,
    class_weight=None,
    n_jobs=-1,
)
lr_final.fit(X_train_bal, y_train_bal)
```

### 3.4 Ensemble (Stacking)

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier

stack_estimators = [
    ("lr", lr_final),
    ("svm", svm_final),
    ("mlp", mlp),
]

meta_clf = LogisticRegression(
    C=1.0,
    solver="lbfgs",
    multi_class="auto",
    max_iter=1000,
    class_weight="balanced",
    n_jobs=-1,
)

stack_clf = StackingClassifier(
    estimators=stack_estimators,
    final_estimator=meta_clf,
    cv=5,
    n_jobs=-1,
    stack_method="auto",
)

stack_clf.fit(X_train, y_train)
```
